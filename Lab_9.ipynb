# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    predecessors = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')
        predecessors[node] = None

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None: # Handles disconnected graphs or unreachable nodes
            break

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node

    return distances, predecessors
```

```python
graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10},
    'C': {'A': 2, 'B': 5, 'D': 3},
    'D': {'B': 10, 'C': 3}
}

distances, predecessors = dijkstra(graph, 'A')

print("Shortest Distances from node A:")
for node, dist in distances.items():
    print(f"  To {node}: {dist}")
```

```python
def reconstruct_path(predecessors, start, end):
    path = []
    current = end
    while current is not None:
        path.insert(0, current)
        current = predecessors[current]

    # If the path doesn't start with the start node, it means there's no path
    if path[0] != start:
        return []
    return path


print("Shortest Paths from node A:")
for node in distances:
    path = reconstruct_path(predecessors, 'A', node)
    if path:
        print(f"  To {node}: {' -> '.join(path)} (Distance: {distances[node]})")
    else:
        print(f"  To {node}: No path found")
```

```python
# Add two more nodes to the graph
graph_expanded = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10},
    'C': {'A': 2, 'B': 5, 'D': 3, 'E': 6},
    'D': {'B': 10, 'C': 3, 'F': 8},
    'E': {'C': 6, 'F': 3},
    'F': {'D': 8, 'E': 3}
}

# Test with a different source node (e.g., 'C')
distances_C, predecessors_C = dijkstra(graph_expanded, 'C')

print("\nShortest Distances from node C:")
for node, dist in distances_C.items():
    print(f"  To {node}: {dist}")

print("\nShortest Paths from node C:")
for node in distances_C:
    path = reconstruct_path(predecessors_C, 'C', node)
    if path:
        print(f"  To {node}: {' -> '.join(path)} (Distance: {distances_C[node]})")
    else:
        print(f"  To {node}: No path found")
```
