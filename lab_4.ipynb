- To implement Linear Search, measure its time complexity for searching an element, vary the input size by changing the number of elements in the list, and plot a graph to visualize the relationship between time taken and input size
- Algorithm:  
1. Start with the first element of the list.
2. Compare the element with the value being searched for.
3. If the element matches, return the index.
4. If the element does not match, move to the next element and repeat steps 2-3.
5. If the end of the list is reached and no match is found, return -1

```python
def linear_search(arr, target):
    """
    Implements the Linear Search algorithm.

    Args:
        arr (list): The list of elements to search through.
        target: The element to search for.

    Returns:
        int: The index of the target element if found, otherwise -1.
    """
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

print("Linear Search function defined.")
```

Algorithm:
1. Start with the middle element of the list
2. If the middle element matches the element being searched, return the index of the middle element
3. If the middle element is greater than the element being searched, recursively call the function on the left half of the list
4. If the middle element  is  less than the element  being searched, recursively call the function on the right half of the list
5. If the element is not found, return -1

```python
#Prob_2: write code for recursive Binary Search and plot a graph to visualize the relationship between time taken and input size
```

```python
def recursive_binary_search(arr, target, low, high):
    """
    Implements the recursive Binary Search algorithm.

    Args:
        arr (list): The sorted list of elements to search through.
        target: The element to search for.
        low (int): The lower index of the current search space.
        high (int): The upper index of the current search space.

    Returns:
        int: The index of the target element if found, otherwise -1.
    """
    if low > high:
        return -1  # Element not found

    mid = (low + high) // 2

    if arr[mid] == target:
        return mid  # Element found at mid index
    elif arr[mid] > target:
        return recursive_binary_search(arr, target, low, mid - 1)  # Search in the left half
    else:
        return recursive_binary_search(arr, target, mid + 1, high)  # Search in the right half

print("Recursive Binary Search function defined.")
```

#Prob-4
Sort a given set of elements using the Insertion sort and determine the time required tosort the elements. Repeat the experiment for different values of n, the number of elements in the list to be sorted and plot a graph of the time taken versus n.  
- Aim :  To Develop a program to implement Insertion sort and determine the time required to sort the elements.  
- Algorithm:  
1. Iterate over the array from the second element to the end (i = 1 to n-1).
2. For each element, compare it to the elements before it (j = i-1 to 0).
3. If the current element is smaller, swap it with the element before it.
4. Repeat step 3 until the current element is no longer smaller than the previous element.
5. Repeat steps 2-4 for all elements in the array.

```python
def insertion_sort(arr):
    """
    Implements the Insertion Sort algorithm.

    Args:
        arr (list): The list of elements to be sorted.
    """
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

print("Insertion Sort function defined.")
```

```python
import time
import random

# Define a list of n values to test
n_values_insertion = [10, 100, 500, 1000, 2000, 5000, 10000]

# Initialize lists to store results
list_sizes_insertion = []
times_taken_insertion = []

print("Starting experiment for Insertion Sort...")

for n in n_values_insertion:
    # a. Generate a list of n random integers (unsorted for sorting algorithm)
    test_list_insertion = [random.randint(0, n * 2) for _ in range(n)]

    # c. Record the start time
    start_time_insertion = time.time()

    # d. Call the insertion_sort function
    insertion_sort(test_list_insertion.copy()) # Use a copy to avoid modifying the original list if needed elsewhere

    # e. Record the end time
    end_time_insertion = time.time()

    # f. Calculate the time difference
    time_diff_insertion = end_time_insertion - start_time_insertion

    # g. Append results
    list_sizes_insertion.append(n)
    times_taken_insertion.append(time_diff_insertion)

print("Experiment complete. Results for Insertion Sort:")
print(f"List Sizes (n): {list_sizes_insertion}")
print(f"Times Taken (seconds): {times_taken_insertion}")
```

## Implement Timing and Data Generation

### Subtask:
Write Python code to generate lists of varying sizes (n), and for each size, measure the time taken by the `linear_search` function to find an element. Store the list sizes and corresponding times.


**Reasoning**:
To measure the execution time of the `linear_search` function for varying input sizes, I will generate random lists, perform the search, and record the time taken, following the given instructions.



```python
import time
import random

# Define a list of n values to test
n_values = [10, 100, 500, 1000, 2000, 5000, 10000]

# Initialize lists to store results
list_sizes = []
times_taken = []

print("Starting experiment for Linear Search...")

for n in n_values:
    # a. Generate a list of n random integers
    test_list = [random.randint(0, n * 2) for _ in range(n)]

    # b. Select a target element to search for (known to be in the list)
    target = test_list[random.randint(0, n - 1)]

    # c. Record the start time
    start_time = time.time()

    # d. Call the linear_search function
    linear_search(test_list, target)

    # e. Record the end time
    end_time = time.time()

    # f. Calculate the time difference
    time_diff = end_time - start_time

    # g. Append results
    list_sizes.append(n)
    times_taken.append(time_diff)

print("Experiment complete. Results:")
print(f"List Sizes (n): {list_sizes}")
print(f"Times Taken (seconds): {times_taken}")
```

**Reasoning**:
The previous step successfully generated data for list sizes and their corresponding linear search times. The next logical step, as per the overall task, is to visualize this data by plotting a graph of time taken versus the number of elements (n).



```python
import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(list_sizes, times_taken, marker='o', linestyle='-', color='b')

# Add labels and title
plt.title('Linear Search Time Complexity')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (seconds)')
plt.grid(True)

# Display the plot
plt.show()
```
